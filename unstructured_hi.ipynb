```python
# Install the unstructured library with all document dependencies
!pip install "unstructured[all-docs]"

# Import the necessary function
from unstructured.partition.auto import partition

# Download a sample document to test with
!wget https://raw.githubusercontent.com/Unstructured-IO/unstructured/main/example-docs/eml/fake-email.eml -O sample.eml

# Partition the document into elements
elements = partition(filename="sample.eml")

# Print the extracted elements
for element in elements:
    print(f"Type: {type(element).__name__}")
    print(f"Text: {element.text[:100]}...")  # Print first 100 chars
    print("-" * 40)
```
